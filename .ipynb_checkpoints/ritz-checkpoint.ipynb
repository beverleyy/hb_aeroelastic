## Mode shape calculation using Rayleigh-Ritz method

This piece of code calculates mode shape of an Airbus A350 wing using Rayleigh-Ritz method with polynomial basis functions (power series). Engine and pylons have been assumed as point masses.

This can be adapted for other wing shapes by changing the wing characteristics like material density, surface area function, inertia function, the engine mass, pylon mass, etc or adding/removing some point masses for better estimates.

```python
import numpy as np
import scipy

import matplotlib.pyplot as plt
```

```python
## Define parameters of wing
L = 29
E = 2e+12
rho = 300

area = lambda y: 4.1*(1.005-y/L)
inertia = lambda y: 0.03*(1.005-0.5*y/L-0.5*((y/L)**2))

pylons = {}
```

```python
## Define functions
L = 29

phi1_dp = lambda y,i: (L^(-3))*(L-y)*((y/L)**(i-1));
phi1 = lambda y,i: ((y/L)**(i+1)*(2+i-i*(y/L)))/(i*(i+1)*(i+2))
```

```python
def solveWing(phi,phi_dp,n):
    E = 2e+12
    L = 29
    rho = 300
    
    area = lambda y: 4.1*(1.005 - y/L)
    inertia = lambda y: 0.03*(1.005-0.5*y/L-0.5*((y/L)**2))
    
    # Pylons
    mp = 300
    yp1 = L/8
    yp2 = 2*L/5
    yp3 = L/2
    
    # Engine
    me = 6000
    ye = L/4
    
    ## Assembly
    M = np.zeros((n,n))
    K = np.zeros((n,n))
    
    for i in range(1,n+1):
        for j in range(1,n+1):
            # Integrate for wing mass
            mass1 = lambda y: rho*area(y)*phi(y,i)*phi(y,j)
            mass_int = scipy.integrate.quad(mass1,0,L)[0]
            
            # Add back the pylon masses
            M[i-1,j-1] = mass_int + mp*(
                phi(yp1,i)*phi(yp1,j) +
                phi(yp2,i)*phi(yp2,j) +
                phi(yp3,i)*phi(yp3,j)
            ) + me*phi(ye,i)*phi(ye,j)
            
            # Stiffness
            stiffness = lambda y: E*inertia(y)*phi_dp(y,i)*phi_dp(y,j)
            K[i-1,j-1] = scipy.integrate.quad(stiffness,0,L)[0]
            
    ## Eigenvalues
    [D,V] = scipy.linalg.eig(K,M)
    idx = np.argsort(D)
    d = D[idx]
    
    vals = np.sqrt(d/(2*np.pi))
    xi = V[:,idx]
    
    print(vals)
        
    return vals,xi
```

```python
## Number of modes
n = 4

## Solve and print the eigenvalues
[vals,xi] = solveWing(phi1,phi1_dp,n)
```

```python
fig,ax = plt.subplots()

ax.axhline(y=0,color='k',linestyle=':',linewidth=1)

# No. of plotting points
N = 100
ysol = np.linspace(0,L,N)
    
# Just plot the first four
for i in range(1,5):
    mode = np.zeros((N))
    
    # Need to sum all the modes
    for k in range(1,n+1):
        mode = mode + xi[k-1,i-1]*phi1(ysol,k)
        
    ax.plot(ysol/L,mode/mode[-1],label='i='+str(i))
    
ax.set_xlim([0,1])
ax.set_xlabel('y/L')

ax.legend()

ax.set_title('First 4 mode shapes')
```
