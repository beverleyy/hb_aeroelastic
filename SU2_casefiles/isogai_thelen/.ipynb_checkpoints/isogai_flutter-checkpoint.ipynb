# Aeroelastic Flutter Prediction with Isogai Aerofoil

This Jupyter notebook employs the harmonic balance method to solve for aerodynamic forces in an aeroelastic solution framework using the P-K method, replicating a case presented in Thelen et al. (2019).

```python
import numpy as np
import scipy
import scipy.linalg
import scipy.optimize

import matplotlib.pyplot as plt
import pandas as pd
```

## Airfoil setup and parameters

The case describes a rigid 2D NACA 64A010 aerofoil with 2 degrees of freedom (plunge, pitch). This is a common benchmark problem studied extensively by Koji Isogai with the aim to compute flutter speed as the Mach number is varied. The equations of motion describing this system are:

$$
\begin{bmatrix}1 & -x_\alpha\\ -x_\alpha & r_\alpha^2\end{bmatrix}\begin{Bmatrix}\ddot{h}\\ \ddot{\alpha}\end{Bmatrix} + \begin{bmatrix}\omega_h^2 & 0 \\ 0 & -r_\alpha^2\omega_\alpha^2\end{bmatrix}
\begin{Bmatrix}h\\ \alpha\end{Bmatrix} = \frac{V_\infty^2\overline{c}S}{2\pi\mu b^2}\begin{Bmatrix}C_l\\C_m\end{Bmatrix}
$$

Here, $x_\alpha = x_{cg}-x_{ea}$ refers to the static imbalance, $r_\alpha$ is the radius of gyration about the elastic axis, $m$ is the aerofoil's mass (normalized with respect to the chord length), and $\{\omega_h,\omega_\alpha\}$ are the natural frequencies of the linear and torsional springs of the aerofoil. Finally, $\mu=m/{\pi\rho_\infty b^2}$ is the aerofoil-to-fluid mass ratio and $V_\infty$ is a variable freestream speed.

```python
# Center of gravity [ft]
x_cg = 0.4

# Elastic axis [ft]
x_ea = -0.5

# Linear and torsional springs [rad/s]
omega_h = 100
omega_a = 100

# Static imbalance [ft]
x_a = x_cg-x_ea

# Radius of gyration [ft]
r = np.sqrt(0.87)

# mu = m/(pi*rho*b**2), airfoil to fluid mass ratio where m is the airfoil mass
mu = 60
```

## Calculating structural properties

Since the system has only 2 degrees of freedom, based on the setup the mass matrix, $\mathbf{M}$, is given by \begin{bmatrix}
1 & -x_a\\
-x_a & r^2
\end{bmatrix} and the stiffness matrix $\mathbf{K}$ is given by \begin{bmatrix}
-\omega_h^2 & 0\\
0 & r^2\omega_a^2
\end{bmatrix}

The mode shapes and natural frequencies of the structure are given by solving the eigenvalue problem $-\omega^2\mathbf{M}+\mathbf{K}=0$, where the natural frequencies are the eigenvalues and the mode shapes are the eigenvectors.

$\phi$ gives a square matrix consisting of mode shape vectors in each column.

```python
## Mass matrix
M = np.array([[1,-x_a],[-x_a,r**2]])
M
```

```python
## Stiffness matrix
K = np.array([[omega_h**2,0],[0,(r*omega_a)**2]])
K
```

```python
## Modes (2 DOF = 2 modes)
[D,V] = scipy.linalg.eig(K,M)
```

```python
# Normalize modes
phi = V/V.max(axis=0)
phi
```

## Calculating generalized masses

The generalized mass for each mode is given by
$$m_i = \mathbf{\phi}_i^T\mathbf{M}\mathbf{\phi}_i$$
and so in this case, the generalized mass matrix, $\overline{\mathbf{M}}$ is \begin{bmatrix}m_1&0\\0&m_2\end{bmatrix}
The generalized stiffness matrix in this 2DOF case is given by $\overline{\mathbf{K}}=$\begin{bmatrix}m_1\omega_1^2&0\\0&m_2\omega_2^2\end{bmatrix}
and likewise as for the generalized mass, the value of $m_i\omega_i^2$ can be calculated using 
$$m_i\omega_i^2 = \mathbf{\phi}_i^T\mathbf{K}\mathbf{\phi}_i$$
Knowing the value of $m_i$, the value of $\omega_i$ can be calculated trivially.

```python
masses = np.zeros(np.size(D))
stiffs = np.zeros(np.size(D))

## Calculate normalized matrices
for i in range(0,np.size(D)):
    masses[i] = np.matmul(phi[:,[i]].T,np.matmul(M,phi[:,[i]]))
    stiffs[i] = np.matmul(phi[:,[i]].T,np.matmul(K,phi[:,[i]]))
    
## Calculate natural freqs
omegas = np.sqrt(stiffs/masses)
```

```python
print(f'Generalized masses are {masses}')
print(f'Natural frequences of each mode are {omegas} rad/s')
```

## Aerodynamics post-processing

The modes calculated above (i.e., $\mathbf{\phi}$) are prescribed to SU2 by way of prescribing the displacements $\Delta x$ and $\Delta z$ to the grid movement directly, and the below code processes the output. Since this is a simple rigid 2DOF system, the displacements can be calculated trivially.

```python
Mach = 0.85
U_inf = Mach*np.sqrt(1.4*287*288.15)
```

### Unsteady solver

To validate the harmonic balance code, a transient Mach=0.85 simulation with reduced frequency $k=0.15$ is run using the SU2 code first for comparison. 

```python
path = 'mach085/unsteady'
```

```python
## Function to get unsteady data from csv file
def get_coeffs(csv_path,Nstart=0):
    '''
    csv_path should be in the format PATH/TO/ENTIRE/CASE/IDENTIFIER
    e.g. if case directories are ~/Documents/SU2/Run/isogai_aeroelas_HB/{1,2,3,4...}/{pitch, plunge},
         then csv_path is '~/Documents/SU2/Run/isogai_aeroelas_HB/IDENTIFIER'
    '''
    mode1df = pd.read_csv(csv_path+'/mode1/history_00002.csv').rename(columns=lambda x: x.strip())
    mode2df = pd.read_csv(csv_path+'/mode2/history_00002.csv').rename(columns=lambda x: x.strip())
    
    times1 = mode1df['"Cur_Time"'][Nstart:].to_numpy()
    times2 = mode2df['"Cur_Time"'][Nstart:].to_numpy()
    
    times1 = times1-times1[0]
    times2 = times2-times2[0]
    
    lift1 = -mode1df['"CL"'][Nstart:].to_numpy()
    lift2 = -mode2df['"CL"'][Nstart:].to_numpy()
    
    moment1 = -mode1df['"CMz"'][Nstart:].to_numpy()
    moment2 = -mode2df['"CMz"'][Nstart:].to_numpy()
    
    return times1,times2,lift1,lift2,moment1,moment2
```

### Harmonic Balance

The same simulation is then repeated using the harmonic balance method.

```python
omega = 0.15*U_inf/0.5
omega
```

```python
hb_path = 'mach085/HB'
```

```python
## Define functions for Einv and interpolation

#  Einv calculation
def get_Einv(T,N,K,omegas):
    Einv = np.zeros((N,K),dtype=complex)
    for n in range(0,N):
        for k in range(0,K):
            Einv[n,k] = complex(np.cos(omegas[k]*n*T/N),np.sin(omegas[k]*n*T/N))
    return Einv

#  Interpolation
def get_interp(T,times,omegas,hb_lift):
    Einv = get_Einv(T,len(omegas),len(omegas),omegas)
    coeffs = np.squeeze(np.asarray(np.mat(np.linalg.inv(Einv))*np.mat(hb_lift).transpose()))
    interpol = np.zeros(len(times),dtype=complex)
    for i in range(0,len(times)):
        for k in range(0,len(coeffs)):
            interpol[i] = interpol[i] + coeffs[k]*complex(np.cos(omegas[k]*times[i]),np.sin(omegas[k]*times[i]))
    return interpol
```

```python
def get_hbcoeffs(csv_path,omega=86.766975381,N=3):
    '''
    csv_path should be in the format PATH/TO/ENTIRE/CASE/IDENTIFIER
    e.g. if case directories are ~/Documents/SU2/Run/isogai_aeroelas_HB/{1,2,3,4...}/{pitch, plunge},
         then csv_path is '~/Documents/SU2/Run/isogai_aeroelas_HB/IDENTIFIER'
    '''
    T = 2*np.pi/omega
    
    hb1_df = pd.read_csv(csv_path+'/mode1/HB_output.csv',index_col=False)
    hb2_df = pd.read_csv(csv_path+'/mode2/HB_output.csv',index_col=False)

    hb1_times = hb1_df['time_instance'].to_numpy()*T*omega/N
    hb1_lift = hb1_df['CL'].to_numpy()
    hb1_moment = hb1_df['CMz'].to_numpy()

    hb2_times = hb2_df['time_instance'].to_numpy()*T*omega/N
    hb2_lift = hb2_df['CL'].to_numpy()
    hb2_moment = hb2_df['CMz'].to_numpy()
    
    return hb1_times,hb1_lift,hb1_moment,hb2_times,hb2_lift,hb2_moment
```

```python
## Function to get data from HB output csv files
def get_hb(csv_path,omega=86.766975381,n_times=100):
    hb1_times,hb1_lift,hb1_moment,hb2_times,hb2_lift,hb2_moment = get_hbcoeffs(csv_path)

    ## Need to grab periods from config files
    T = 2*np.pi/omega

    ## Need to grab frequencies from config file
    omegas = [0.,omega,-omega]
    
    ## Number of time points to interpolate
    times = np.linspace(0,2*T,num=n_times)

    ## Get interpolated values
    lift1 = -get_interp(T,times,omegas,hb1_lift)
    moment1 = -get_interp(T,times,omegas,hb1_moment)

    lift2 = -get_interp(T,times,omegas,hb2_lift)
    moment2 = -get_interp(T,times,omegas,hb2_moment)
    
    return times,lift1,lift2,moment1,moment2
```

### Generalized Aerodynamic Forces (GAF)

From Thelen et al. (2019), these are given by $\phi^T\mathbf{f}$, i.e., for Mode 1, the GAF is calculated with $F = \phi_1^T\begin{bmatrix}C_L \\ C_D\end{bmatrix}$

```python
def get_GAFs(lift1,lift2,moment1,moment2):
    ## GAFs from Mode 1 excitation
    f11 = phi.T[0][0]*lift1 + phi.T[0][1]*moment1 #inf on mode 1
    f12 = phi.T[1][0]*lift1 + phi.T[1][1]*moment1 #inf on mode 2

    ## GAFs from Mode 2 excitation
    f21 = phi.T[0][0]*lift2 + phi.T[0][1]*moment2 #inf on mode 1
    f22 = phi.T[1][0]*lift2 + phi.T[1][1]*moment2 #inf on mode 2
    
    return f11.real,f12.real,f21.real,f22.real
```

### Generalized Aerodynamic Influence Coefficients (GAIC)

Using sinusoidal fitting to obtain complex number magnitude/phase for each GAF, then divide by displacement (1e-4).

```python
def fit_sin(tt, yy):
    '''Fit sin to the input time sequence, and return fitting parameters "amp", "omega", "phase", "offset", "freq", "period" and "fitfunc"'''
    tt = np.array(tt)
    yy = np.array(yy)
    ff = np.fft.fftfreq(len(tt), (tt[1]-tt[0]))   # assume uniform spacing
    Fyy = abs(np.fft.fft(yy))
    guess_freq = abs(ff[np.argmax(Fyy[1:])+1])   # excluding the zero frequency "peak", which is related to offset
    guess_amp = np.std(yy) * 2.**0.5
    guess_offset = np.mean(yy)
    guess = np.array([guess_amp, 2.*np.pi*guess_freq, 0., guess_offset])

    def sinfunc(t, A, w, p, c):  return A * np.sin(w*t + p) + c
    popt, pcov = scipy.optimize.curve_fit(sinfunc, tt, yy, p0=guess)
    A, w, p, c = popt
    f = w/(2.*np.pi)
    fitfunc = lambda t: A * np.sin(w*t + p) + c
    return {"amp": A, "omega": w, "phase": p, "offset": c, "freq": f, "period": 1./f, "fitfunc": fitfunc, "maxcov": np.max(pcov), "rawres": (guess,popt,pcov)}
```

```python
def get_GAIC(times1,times2,f11,f12,f21,f22,displ=0.0001j):
    fm11 = f11-np.mean(f11)
    fm12 = f12-np.mean(f12)
    fm21 = f21-np.mean(f21)
    fm22 = f22-np.mean(f22)
    
    res11 = fit_sin(times1, np.float64(fm11))
    res12 = fit_sin(times1, np.float64(fm12))
    res21 = fit_sin(times2, np.float64(fm21))
    res22 = fit_sin(times2, np.float64(fm22))
    
    fit11 = res11['amp']*(np.cos(res11['phase'])+1j*np.sin(res11['phase']))
    fit12 = res12['amp']*(np.cos(res12['phase'])+1j*np.sin(res12['phase']))
    fit21 = res21['amp']*(np.cos(res21['phase'])+1j*np.sin(res21['phase']))
    fit22 = res22['amp']*(np.cos(res22['phase'])+1j*np.sin(res22['phase']))
    
#     fit11 = res11['amp']*np.exp(1j*res11['phase'])
#     fit12 = res12['amp']*np.exp(1j*res12['phase'])
#     fit21 = res21['amp']*np.exp(1j*res21['phase'])
#     fit22 = res22['amp']*np.exp(1j*res22['phase'])
    
    Q = np.array([
        [fit11, fit21],
        [fit12, fit22]
    ])/(-displ)
    
    return Q
```

## Applied Aeroelasticity

Here we calculate the generalized aerodynamic influence coefficients (GAIC) for Mach 0.875 and $k_{red} = [0.05, 0.4]$ at intervals of 0.05 to perform flutter analysis for the Mach 0.875 case.

```python
multipath = 'mach0875/'
```

```python
k_red = np.linspace(0.05,0.4,8)
b = 0.5

Mach = 0.875
U_inf = Mach*np.sqrt(1.4*287*288.15)
```

`q_set` contains all the 2x2 GAIC matrices computed from simulations at the different values of $k_{red}$.

```python
q_set = []
for i in range(len(k_red)):
    w = k_red[i]*U_inf/b
    times,hbl1,hbl2,hbm1,hbm2 = get_hb(multipath+str(i+1),omega=w)
    hf11,hf12,hf21,hf22 = get_GAFs(hbl1,hbl2,hbm1,hbm2)
    q = get_GAIC(times,times,hf22,hf21,hf12,hf11,0.0001j)
    q_set.append(q)
q_set
```

We need to create a new bunch of Q(k) for numerical robustness and smoothness. Here we use PCHIP module in Scipy as per Thelen et al. (2019).

```python
from scipy.interpolate import pchip_interpolate

k_expanded = np.linspace(min(k_red),max(k_red),num=400)
q_expanded = pchip_interpolate(k_red,q_set,k_expanded)
```

```python
## Need to subset the terms so that the plotter actually works
def sortQs(q_set):
    q11 = []
    q12 = []
    q21 = []
    q22 = []
    for i in range(len(q_set)):
        q11.append(q_set[i][0,0])
        q12.append(q_set[i][0,1])
        q21.append(q_set[i][1,0])
        q22.append(q_set[i][1,1])
    return q11,q12,q21,q22
```

Below are plots of how the magnitude of GAICs changes with respect to the reduced frequency at Mach 0.875.

```python
## Plot
q11,q12,q21,q22 = sortQs(q_set)
interq11,interq12,interq21,interq22 = sortQs(q_expanded)

fig,ax = plt.subplots(2,2)
fig.set_size_inches([10,6])
fig.tight_layout(rect=(0, 0, 0.85, 1))

ax[0,0].plot(k_red,np.abs(q11),"o",label="HB")
ax[0,0].plot(k_expanded,np.abs(interq11),label="pchip")

ax[0,1].plot(k_red,np.abs(q12),"o",label="HB")
ax[0,1].plot(k_expanded,np.abs(interq12),label="pchip")

ax[1,0].plot(k_red,np.abs(q21),"o",label="HB")
ax[1,0].plot(k_expanded,np.abs(interq21),label="pchip")

ax[1,1].plot(k_red,np.abs(q22),"o",label="HB")
ax[1,1].plot(k_expanded,np.abs(interq22),label="pchip")

#ax[0,1].legend()

for a in ax:
    for aa in a:
        aa.set_xlabel('k')
        
ax[0,0].set_ylabel('$Q_{11}$')
ax[0,1].set_ylabel('$Q_{12}$')
ax[1,0].set_ylabel('$Q_{21}$')
ax[1,1].set_ylabel('$Q_{22}$')

lines_labels = [ax[0,0].get_legend_handles_labels()] #for ax in fig.axes]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, loc='center right')#, ncol=len(labels))
```

## P-K calculation

Finally, the repository of GAICs (1 2x2 matrix for each reduced frequency) for Mach 0.875 are coupled with the structural solver and the flutter speed is searched within a large range of freestream velocities. The aeroelastic methodology used here is the P-K method which is chosen for its fast calculation time compared to others.

```python
## Calculate normalized mass and stiffness matrices (same as above)
Mbar = np.matmul(phi.T,np.matmul(M,phi))
Kbar = np.matmul(phi.T,np.matmul(K,phi))
```

```python
u = np.arange(40,400,1)

## M = u/sqrt(gamma*R*T) = constant
## sqrt(gamma*R*T) = u/M
## gamma*R*T = (u/M)**2
## T = ((u/M)**2)/(gamma*R)
```

```python
## Make empty storage arrays
pkpitch = []
pkheave = []
pkpdamp = []
pkhdamp = []
```

```python
## Loop through each airspeed
for i in range(len(u)):
    
    U = u[i]
    K = (b/U)*omegas #reduced frequency
    Mhh = Mbar
    
    rho = 4/(np.pi*mu); #SI units
    q_inf = 0.5*rho*U*U
    
    ## Select the nth degree of freedom
    for n in range(len(K)):
        k = K[n]
        U = u[i]
        Mhh = Mbar
        
        # Start while loop for convergence
        omega_old = omegas[n]
        conv = 1e-6
        err = 1
        count = 1
        
        while abs(err)>conv:
            Q = pchip_interpolate(k_red,q_set,k)
            
            # Calculate matrices
            Bhh = (-q_inf*b/U/k)*Q.imag
            Khh = Kbar - q_inf*Q.real

            # Setup
            PK_A = np.concatenate((
                np.concatenate((np.zeros((2,2)),np.eye(2)),axis=1),
                np.concatenate((-np.linalg.solve(Mhh,Khh),-np.linalg.solve(Mhh,Bhh)),axis=1)
            ),axis=0)

            # Solve
            PKvals = np.linalg.eigvals(PK_A)
            PKvals = sorted(PKvals,key=lambda x:x.imag,reverse=True)
            
            omega_new = PKvals[n].imag
            
            if(omega_new):
                err = abs(abs(omega_old)-abs(omega_new))/omega_old
                k = omega_new*b/U
                omega_old = omega_new
                count = count+1
            else:
                break        
        #end while
        
        ## THE FREQUENCIES ARE IN RADIANS
        if n==0:
            pkheave.append(abs(PKvals[n]))
            pkhdamp.append(PKvals[n].real/abs(PKvals[n]))
        else:
            pkpitch.append(abs(PKvals[n]))
            pkpdamp.append(PKvals[n].real/abs(PKvals[n]))
            
        ## Flip values if wrong
        if n:
            if abs(pkpitch[i]-pkpitch[i-1])>0.1:
                pkpitch[i] = abs(PKvals[not n])
                pkpdamp[i] = PKvals[not n].real/abs(PKvals[not n])
            if abs(pkheave[i]-pkheave[i-1])>0.1:
                pkheave[i] = abs(PKvals[not n])
                pkhdamp[i] = PKvals[not n].real/abs(PKvals[not n])
```

```python
def check_flutter(damping2,damping1):
    c = 0    #c=1 indicates pitch mode flutter
    idx = 0
    for n in range(len(damping1)-1):
        z_next = damping1[n+1]
        z_this = damping1[n]
        if abs(z_this)>1e-12:
            if z_next/z_this<0:
                idx = n+1
                c = 1
                break
    if c==0:
        for n in range(len(damping2)-1):
            z_next = damping2[n+1]
            z_this = damping2[n]
            if abs(z_this)>1e-12:
                if z_next/z_this<0:
                    idx = n+1
                    break
    return idx,c
```

```python
idx,c = check_flutter(pkpdamp,pkhdamp)
if idx:
    pkfspd = u[idx]
    if c:
        pkfreq = pkpitch[idx]
    else:
        pkfreq = pkheave[idx]
    print(f'Test frequency is {omegas[1]/2/np.pi} Hz')
    print(f'Unsteady flutter speed is {pkfspd} m/s')
    print(f'Unsteady flutter freq. is {pkfreq} Hz')
else:
    print(f'Test frequency is {omegas[1]/2/np.pi} Hz')
    print(f'No flutter observed in range')
```
